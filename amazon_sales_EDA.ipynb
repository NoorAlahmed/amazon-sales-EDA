## processsing and cleaning the dataset

```python
# Changing the data type of discounted price and actual price

df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
# Changing Datatype and values in Discount Percentage

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
# Finding unusual string in rating column
df['rating'].value_counts()
```

```python
# Check the strange row
df.query('rating == "|"')
```

```python
# Changing Rating Columns Data Type

df['rating'] = df['rating'].str.replace('|', '3.9').astype('float64')
# replaced it with 3.9 as it was in the orginal file
```

```python
# Changing 'rating_count' Column Data Type

df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

handling the missing values

```python
# make figure size
plt.figure(figsize=(22, 10))

# count missing values
missing_values = df.isnull().sum()

# plot counts
missing_values.plot(kind='bar')

# add labels
plt.xlabel('Columns')
plt.ylabel('Count')
plt.title('Count of Missing Values in Each Column')

plt.show()
```

```python
#display rows where the 'rating_count' column has missing values
df[df['rating_count'].isnull()].head(5)
```

```python
# Impute missing values
df['rating_count'] = df.rating_count.fillna(value=df['rating_count'].median())
```

## Visualization

```python
plt.figure()
plt.scatter(df['discounted_price'], df['rating'])
plt.title("Price vs Rating")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()

"""This scatter plot shows the relationship between product price and rating.
It helps determine whether higher prices lead to better ratings."""
```

```python
plt.figure()
plt.hist(df['rating'], bins=20)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
"""This histogram shows the distribution of product ratings in the dataset.
It helps us understand how customers are rating products overall and identify any trends"""
```

```python
plt.figure()
plt.scatter(df['discount_percentage'], df['rating'])
plt.title("Discount vs Rating")
plt.xlabel("Discount %")
plt.ylabel("Rating")
plt.show()
"""This scatter plot shows the relationship between discount percentages and product ratings
It helps] understand whether higher discounts are associated with higher or lower customer ratings"""
```

```python
avg_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False)

plt.figure()
avg_rating.head(10).plot(kind='bar')
plt.title("Average Rating per Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
"""This bar chart shows the average rating for each product category
It helps us identify which categories have the highest or lowest customer satisfaction overall"""
```

```python
top_products = df.sort_values(by='rating_count', ascending=False).head(10)

plt.figure()
plt.barh(top_products['product_name'], top_products['rating_count'])
plt.title("Top Reviewed Products")
plt.xlabel("Number of Reviews")
plt.ylabel("Product")
plt.show()
"""This horizontal bar chart shows the top 10 products based on the number of reviews
It helps us identify which products are the most popular and heavily reviewed by customers"""
```

```python
plt.figure()
plt.hist(df['discounted_price'], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
"""This histogram shows the distribution of discounted prices in the dataset
It helps us understand the common price ranges and trends in product pricing"""
```
